1. Select all the rows from the given dataframe in which 'Percentage' is greater than 80 using basic method.

```
import pandas as pd
import numpy as np

data = {'Name': ['Amit', 'Keerthy', 'Prem', 'Priya','Pragna'],
        'Percentage': [88, 92, 95, 70, 65]}
df = pd.DataFrame(data)
selected_df = df[df['Percentage'] > 80]
selected_df
```

2. Select all the rows from the given dataframe in which ‘Percentage’ is greater than 80 using loc.


```
import pandas as pd
import numpy as np

data = {'Name': ['Amit', 'Keerthy', 'Prem', 'Priya','Pragna'],
        'Percentage': [88, 92, 95, 70, 65]}
df = pd.DataFrame(data)

selected_rows = df.loc[df['Percentage'] > 80]
selected_rows
```

3. Display only the 'Name' and 'Percentage' columns using the .loc method.

```
print(df.loc[:, ['Name', 'Percentage']])
```

Display only the 'Name' and 'Percentage' columns for rows where the 'Percentage' is greater than 80 using the .loc method.


```
print(df.loc[df['Percentage'] > 80, ['Name', 'Percentage']])
```

 Update the 'Percentage' value to 90 for the row where the 'Name' is 'Ajay'.


```
import pandas as pd
import numpy as np
data = {'Name': ['Amit', 'Keerthy', 'Prem', 'Priya','Pragna', 'Ajay'],
        'Percentage': [88, 92, 95, 70, 65, 80]}
df = pd.DataFrame(data)
df.loc[df['Name'] == 'Ajay', 'Percentage'] = 90
df
```

Select all the rows where the 'Percentage' is greater than 90 and the 'Stream' is 'Science'

```
data = {'Name': ['Amit', 'Keerthy', 'Prem', 'Priya','Pragna'],
        'Percentage': [88, 92, 95, 70, 65],
        'Stream': ['Commerce', 'Science', 'Science', 'Arts', 'Science']}
df = pd.DataFrame(data)

selected_rows = df[(df['Percentage'] > 90) & (df['Stream'] == 'Science')]
selected_rows
```

Display all the rows where the 'Stream' is not 'Math'

```
data = {'Name': ['Amit', 'Keerthy', 'Prem', 'Priya','Pragna'],
        'Percentage': [88, 92, 95, 70, 65],
        'Stream': ['Commerce', 'Science', 'Science', 'Arts', 'Science']}
df = pd.DataFrame(data)

selected_rows = df[df['Stream'] != 'Math']
selected_rows
```

Sort the rows of the DataFrame in ascending order based on the values in the 'Stream' column.

```
data = {'Name': ['Amit', 'Keerthy', 'Prem', 'Priya','Pragna'],
        'Percentage': [88, 92, 95, 70, 65],
        'Stream': ['Commerce', 'Science', 'Science', 'Arts', 'Science']}
df = pd.DataFrame(data)

sorted_df = df.sort_values(by='Stream', ascending=True)
sorted_df
```
