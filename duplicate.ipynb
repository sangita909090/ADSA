```python
import pandas as pd
data={"Name":["Alice","Bob","Alice","David","Alice"],
      "Age":[25,30,25,40,25],
      "City":["NY","LA","NY","Chicago","AM"]}

df=pd.DataFrame(data)
print(df)
```

#removing duplicates


```python
unique_df=df.drop_duplicates()
display(unique_df)
```

#Drop Duplicates


```python
result=df.drop_duplicates(subset=['Name'])
display(result)
```

#keep the last occurence of duplicates

```python
result=df.drop_duplicates(keep='last')
display(result)
```

```python
#drop all duplicates
result=df.drop_duplicates(keep=False)
display(result)
```

```python
#modify the dataFrame in place
df.drop_duplicates(inplace=True)
display(df)
```
